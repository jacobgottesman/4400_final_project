```python
from utils.preprocessing import *
import pandas as pd
```

```python

def is_valid_tuple_list(item):
    if not isinstance(item, list):
        return False
    if len(item) != 500:
        return False
    for t in item:
        if not (isinstance(t, tuple) and len(t) == 2):
            return False
        if not all(isinstance(x, int) for x in t):
            return False
    return True
```

```python
df = pd.read_csv('data/processed_routes_0_25000.csv')
df['route_xy'] = df['route_xy'].apply(eval)
df = df[df['route_xy'].apply(is_valid_tuple_list)]
# df = df.reset_index(drop=False, names='index')

```

```python
df_50 = pd.read_csv('data/processed_routes_50000_75000.csv')
df_50.index = df_50.index + 50000
df_50['route_xy'] = df_50['route_xy'].apply(eval)
df_50 = df_50[df_50['route_xy'].apply(is_valid_tuple_list)]

df_50.head()
```

```python
both_dfs = pd.concat([df, df_50])
both_dfs = both_dfs.reset_index(drop=False, names='index')
both_dfs
```

```python
both_dfs.to_csv('data/processed_combined.csv')
```
